### Ceteris Paribus Profiles *(what-ifs)*
First we will plot the Cetreris Paribus Profiles for given person. That way we will check what would be model prediction if we changed wan input variable and left the rest unchanged.

```python
def predict_with_changed(column, values, instance, model=trained):
    """ Calculates predictions for given instance with changed values of column.
        Uses provided values and 
    """
    spectrum = []
    for i in range(len(values)):
        obj = instance.copy()
        obj[column] = values[i]
        spectrum.append(obj)
    return model.predict_proba(spectrum)
```

```python
def plot_cpp(column, values, instance, model=trained,title=None, x_values=None, hlines=None):
    # setting default values
    if title is None:
        title = 'Changing column "' + column
        if x_values is None:
            title += '" (' + str(instance[column]) + ')'
        else:
            title += '" (' + str(x_values[int(instance[column])]) + ')'
    if x_values is None:
        x_values = values
    if hlines is None:
        hlines = [0.5]
        
    # getting predictions for changed values
    spectrum  = predict_with_changed(column, values, instance, model)
    real_pred = trained.predict_proba([instance])[0][1]
    real_val  = instance[column]
    
    # plotting results
    plt.ylim(0,1)
    plt.title(title)
    plt.hlines(hlines, values[0], values[-1], '0.5', 'dashed')
    plt.plot(x_values, spectrum[:,1])
    plt.plot(real_val, real_pred, 'ro')
    
```

```python
def plot_person(index):
    person = data.loc[index]
    parsed_person = data.loc[index].copy()
    parsed_person['pred_merried'] = trained.predict([data.loc[index][use_columns]])[0]
    parsed_person['pred_proba_meried'] = trained.predict_proba([data.loc[index][use_columns]])[0][1]
    
    input_person = parsed_person[use_columns]
    
    plt.figure(figsize=(20,10))
    plt.suptitle('Ceteris Paribus Profiles (' + parsed_person['merried'] + ')', fontsize=18)

    plt.subplot(2, 2, 1)
    plot_cpp('age', range(18, 90), instance=input_person)

    plt.subplot(2, 2, 2)
    plot_cpp('hausehold_income_val', list(income_conv.values()), instance=input_person)

    plt.subplot(2, 2, 3)
    plot_cpp('religion_val', list(religion_conv.values()),x_values=list(religion_conv.keys()), instance=input_person)

    plt.subplot(2, 2, 4)
    hs_spectrum = predict_with_changed('same_high_school_val',[0,1], instance=input_person)
    plt.ylim(0,1)
    plt.title("changing high school (" + str(parsed_person["same_high_school"]) + ")")
    plt.hlines(0.5, -1, 2, '0.5', 'dashed')
    bars = plt.bar(["Different high school or NaN","Same high school"], hs_spectrum[:,1])
    bars[int(input_person["same_high_school_val"])].set_color('r')
    
```

## Acumulated Local Effect
### ALE for age

```python
# for age specifing buckets of 5 years
def ale_for_age(model, X):
    age_bucket_len = 5
    age_buckets = range(18,90,age_bucket_len)
    ale_x = []
    ale_y = []
    for bb in age_buckets:
        instances = (X[X['age'].between(bb, bb + age_bucket_len, inclusive=True)]).copy()#  bb + 5 >= age >= bb 
        instances["age"] = bb
        pred_b = model.predict_proba(instances)[:,1] # predict instances with changed age to bb
        if len(ale_y) == 0:
            ale_y.append(sum(pred_b) / len(instances))
            ale_x.append(bb)
        instances["age"] = bb + age_bucket_len
        pred_e = model.predict_proba(instances)[:,1]
        ale = sum(pred_e - pred_b) / len(instances)
        ale_y.append(ale + ale_y[-1])
        ale_x.append(bb + age_bucket_len / 2)
    return ale_x, ale_y
ale_x, ale_y = ale_for_age(trained, test_X)
plt.title("ALE plot for age")
plt.ylim(0,1)
plt.plot(ale_x, ale_y)
```

### ALE for income

```python
# for income we will use specified categories
def ale_for_income(model, X):
    i_ale_x = [2.5, 5, 7.5, 10, 12.5, 15, 20, 25, 30, 35, 40, 50, 60, 75, 85, 100, 125, 150, 175, 200, 250]
    i_ale_y = []
    income_buckets = [2.5, 6.25, 8.75, 11.25, 13.75, 17.5, 22.5, 27.5, 32.5, 37.5, 45, 55, 67.5, 80, 92.5, 112.5, 137.5, 162.5, 182.5, 225, 275]
    for i in range(len(income_buckets) - 1):
        bb = income_buckets[i]
        be = income_buckets[i+1]
        instances = (X[X['hausehold_income_val'].between(bb, be, inclusive=True)]).copy()#  bb + 5 >= age >= bb 
        instances["hausehold_income_val"] = bb
        pred_b = model.predict_proba(instances)[:,1] # predict instances with changed age to bb
        if len(i_ale_y) == 0:
            i_ale_y.append(sum(pred_b) / len(instances))
        instances["hausehold_income_val"] = be
        pred_e = model.predict_proba(instances)[:,1]
        ale = sum(pred_e - pred_b) / len(instances)
        i_ale_y.append(ale + i_ale_y[-1])
    return i_ale_x, i_ale_y
i_ale_x, i_ale_y = ale_for_income(trained, test_X)
plt.title("ALE plot for income")
plt.ylim(0,1)
plt.plot(i_ale_x, i_ale_y)
```

## Comparision of variable impact

```python
def build_up_impact(columns, instance, dataset, model):
    """ Calculates predictions fthe dataset iterativly setting columns values to the instance's one """
    ds = dataset.copy()
    answers = []
    answers.append(model.predict_proba(ds))
    for col in columns:
        ds[col] = instance[col]
        answers.append(model.predict_proba(ds))
    return answers
```

```python
def get_order_var_impact(columns, instance, dataset, model):
    acc = list(map(lambda proba: np.mean(proba[:,1]),build_up_impact(columns, instance, dataset, model)))
    mean = acc[0]
    impacts = { columns[i]: acc[i + 1] - acc[i] for i in range(len(columns)) }
    return (mean, impacts)
```

```python
from itertools import permutations
def shapley_impact(columns, instance, dataset, model):
    impacts = { col: 0 for col in columns}
    perms = list(permutations(columns))
    for order in perms:
        _, imp = get_order_var_impact(order, instance, dataset, model)
        for col in columns:
            impacts[col] += imp[col]
    return { k: v / len(perms) for k, v in impacts.items() }
```

```python
def plot_instance(instance, orders, dataset, model):
    n_cols = len(orders[0])
    cols_range = list(range(n_cols))
    global_mean = -1
    plt.figure(figsize=(15,5))
    for i in range(len(orders)):
        plt.subplot(1, len(orders) + 1, i + 1)
        mean, impacts = get_order_var_impact(orders[i], instance, dataset, model)
        global_mean = mean
        bottoms = [mean]
        for col in orders[i][:-1]:
            bottoms.append(bottoms[-1] + impacts[col])
        plt.title('Order ' + str(i + 1) + ' impact')
        plt.ylim(0,1)
        bars = plt.bar(
            x=cols_range,
            height=[float (impacts[col]) for col in orders[i]], 
            bottom=bottoms, 
            color=['red' if impacts[col] < 0 else 'green' for col in orders[i]],
            tick_label=[ short_names[col] for col in orders[i] ]
        )
        for j in cols_range:
            bars[j].set_hatch(hatch_map[orders[i][j]])
    shapley = shapley_impact(orders[0], instance, dataset, model)
    plt.subplot(1, len(orders) + 1, len(orders) + 1)
    plt.title('Average impact')
    plt.ylim(0,1)
    bars = plt.bar(
        x=cols_range,
        height=[float (shapley[col]) for col in orders[0]], 
        bottom=global_mean, 
        color=['red' if impacts[col] < 0 else 'green' for col in orders[0]],
        tick_label=[ short_names[col] for col in orders[0] ]
    )
    for j in cols_range:
        bars[j].set_hatch(hatch_map[orders[0][j]])

def plt_ith_person_impact(i):
    plot_instance(test_X.loc[persons[i]], columns_orders, train_X, trained)
    plt.suptitle('Person ' + str(i + 1), fontsize=18)
```

```python
def normal_dist_prox(std):
    return lambda a,b: np.exp(- (a-b)**2 / (2 * std**2))

age_proximity = normal_dist_prox(7.5)
religin_proximity = normal_dist_prox(1.25)    
income_proximity = normal_dist_prox(20)
hs_proximity = lambda a, b: 1 - np.abs(a - b)

proximity_functions = (('age', age_proximity), ('religion_val', religin_proximity), ('hausehold_income_val', income_proximity), ('same_high_school_val', hs_proximity))
```

```python
from scipy import stats
def create_random_dataset(size, columns, normal_dist_parameters, ranges):
    m = 30
    sd = 5
    data = {}
    for i in range(len(columns)):
        beg, end = ranges[i]
        if normal_dist_parameters[i]:
            loc, scale = normal_dist_parameters[i]
            s = pd.Series(np.rint(stats.truncnorm.rvs(
                beg - loc, end - loc, loc=loc, scale=scale, size=size)))
        else:
            s = pd.Series(np.random.randint(beg, end+1, size))
        data[columns[i]] = s.fillna(0)
            
    return pd.DataFrame(data)
```

```python
def create_dataset_around(instance, size):
    cols = ['age', 'religion_val',  'same_high_school_val', 'hausehold_income_val']
    normal_params = (instance['age'], 10), (instance['religion_val'], 1.5), None, (instance['hausehold_income_val'], 30)
    ranges = (18, 90), (0, 5), (0, 1), (0, 300)
    return create_random_dataset(size, cols, normal_params, ranges)
```

```python
def samples_proximity(a, b, proximity_functions=proximity_functions):
    proximity = 0
    for (label, f) in proximity_functions:
        proximity += f(a[label], b[label])
    return proximity / len(proximity_functions)
```

```python
def get_proximities(instance, ds, proximity_function=samples_proximity):
    prox = []
    for index, row in ds.iterrows():
        prox.append(proximity_function(instance, row))
    return prox
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from graphviz import Source
from sklearn import tree
import re

def draw_lime_tree(instance, dataset, trained):
    simple_model = DecisionTreeClassifier(random_state=127, max_leaf_nodes=8)
    simple_model.fit(dataset, trained.predict(dataset), sample_weight=get_proximities(instance, dataset))
    path = list(map(str, simple_model.decision_path([instance]).indices))
    model_graph = tree.export_graphviz( simple_model, 
                                        out_file=None, 
                                        filled=True, 
                                        feature_names=test_X.columns, 
                                        class_names=('Not married', 'married',),
                                        rotate=True,
                                        rounded=True,
                                        proportion=True)
    graph_with_path = \
        ';'.join(
            map(lambda x: x if re.sub('^\\n(\d+) .*$', '\g<1>', x) not in path 
                            else re.sub('^\\n(\d+) \[','\\n\g<1> [penwidth="5px", ',x), 
                model_graph.split(';')))
    return Source(graph_with_path)
```

Ogólem można zauważyć, że drzewa uczone na prawdziwym zbiorze danych są dużo bardziej do siebie podobne niż te z losowo generowanych punktów wokół osoby.

Można tam zauważyć parę prawidłowości widocznych również na wykresach *ALE* z drugiego zadania:
- wiek $31.5$, pojawiający się na pierwszym lub drugim poziomie to granica powyżej której wiek tak samo, pozytywnie, wpływa na predykcje
- dochód $41.25$ to pewne wypłaszczenie w wykresach *ALE*

Wyraźnie widać też punkt graniczny przy religii $3.5$ co odpowiada wartości pomiędzy *Once or twice a month* a *Once a week*. Wydaje się to sensowne, gdyż jest to podział na osoby będące zawsze na obowiązkowej mszy, a takie, które przychodzą czasami, rzadko lub nigdy.

Jako, że generalnie drzewa z całego zbioru danych są do siebie bardziej podobne, można łatwo wskazać na różnice. Widać na przykład, że osoba 3 ma w pierwszym węźle dochód, co zgadza się z tym, że w wykresach z poprzedniej pracy domowej u niej dochód wpływał dużo bardziej na predykcje niż w przypadku innych osób.

Bliższe przyjrzenie się tym drzewom pokazuje, że są prawie identyczne, co skłania do wniosku, że przybliżają ogólne działanie modelu, a nie lokalne efekty.

### Generowany zbiór
Patrząc na osoby młodsze (1 i 2) widać, że model jest bardzo skupiony na wieku, tworząc tak szczegółowe zasady jak jeśli masz mniej niż 27 lat i więcej niż 25 (co daje wiek dokładnie 26 lat). W obu tych modelach pojawia się, że jeśli osoba ma mniej niż 26 lat nie jest po ślubie (ze 100% pewnością). Warto zauważyć, że jako iż osoba 1 ma skrajną wartość wieku, większość danych wokół niej (63%) mieściło się w tym przedziale wieku. Oznacza to, że większość drzewa jest wyuczona na pozostałych 400 punktach (co wyjaśnia np. klasę osób w wieku 26 lat). Istotną rolę wieku dla osoby pierwszej widać też w wykresach z poprzedniego zadania.

U praktycznie wszystkich z osób pojawia się, tak jak w modelu z danych, poziom religii $3.5$, co wskazuje na istotną rolę religijności dla całego zbioru danych. W zbiorze wokół osoby 1 religia wydaje się być drugim co do ważności czynnikiem, dalej uczęszczanie do tego samego liceum, a na końcu dochód (nie widać tego w wykresie z poprzedniej pracy domowej dla osoby lecz jest to związane  z dominacją wieku w przewidywaniach i tym, że nie była w grupie osób z religijnością 4 i 5, więc nie był to istotny czynnik).

U osoby 2 (mającej 30 lat, a więc przy magicznej granicy, która oddziela ludzi raczej przed i raczej po ślubie), widać też ważną rolę czynnika dochodów. Osoby o wysokim dochodzie gospodarstwa domowego muszą być młode (mieć mniej niż 30 lat) i niereligijne, żeby model zaklasyfikował je jako nie żonate/zamężne.Osoby powyżej 30 roku życia (do około 40 biorąc pod uwagę jak losowane były dane) są już raczej po ślubie, lecz brane pod uwagę są także wspólne liceum i religijność.

U starszych osób (osoby 3 i 4,  51 i 75 lat) jako ważniejszy w drzewie pojawia się czynnik tego samego liceum. Nie jest to zaskakujące biorąc pod uwagę, że jeśli ma się te 40 lat lub więcej i jest się z osobą znaną z liceum (a więc sprzed ponad 20 lat), to prawdopodobnie jest się z nią w związku małżeńskim (swoją drogą jakieś 20 lat temu więcej nastolatek zachodziło w ciążę, co może też mieć jakieś znaczenie). Widać też próg dochodu, który jest jednak znacząco mniejszy niż u osób 1 i 2, lecz może to być spowodowane tym, że osoby te miały mniejszy dochód, a zbiór danych był generowany wokół nich. Z drugiej strony podział z tym niskim dochodem wciąż pozwala na dobre różnicowanie, co może wskazywać, że wraz z wiekiem dochód ma mniejsze znaczenie dla klasyfikacji. Do podobnego wniosku skłania fakt że na drzewie wokół osoby 3, osoby o dochodzie pomiędzy 17 a 42 tysiące dolarów rocznie  są dalej różnicowane względem wieku, a starsze osoby klasyfikowane są jako po ślubie. 

Powyższe wnioski odnoszą się również do osoby 4 i można to interpretować razem z warościami z poprzedniego zadania. Widać, że gdy dochód był brany pod uwagę jako pierwszy, miał większy negatywny wpływ niż gdy był brany po wieku (czy od wpływu z wartości Shapley'a), można więc wnioskować, że wiek moderuje wpływ dochodu na predykcję. Widać tam też (przy wartościach Shapleya), że religijność osoby 4 również ma znaczenie i zalezność ta jest oddana również na drzewie, gdzie dla dochodów osoby kolejnym kryterium po wieku jest również religia.

```python
from sklearn.metrics import log_loss
```

```python
def feature_permutaion_loss(model, column, X, y, samples=1):
    shuffled = X[column].copy().values
    data = X.drop(column, axis=1)
    data[column] = pd.Series(shuffled, index=X.index)
    loss = 0
    for i in range(samples):
        np.random.shuffle(shuffled)
        loss += log_loss(y.values, model.predict(data)) / samples
    return loss
```

```python
def features_importance(model, X, y, sampels=1, plot=False):
    loss = log_loss(y.values, model.predict(X))
    columns_list = list(X.columns)
    importance = [feature_permutaion_loss(model, column, X, y, sampels) - loss for column in columns_list] 
    if plot:
        plt.bar(columns, importance)
    return loss, importance
```

```python
def multiseries_horizontal_barchart(labels, data, series_names=None, bottoms=None, colors=None, 
                                    barh=0.1, bar_dist=0.025, cols_dist=.2):
    if series_names is None:
        series_names = ['Series ' + str(i) for i in range(len(data))]
    if colors is None:
        colors = [plt.cm.Set2(i) for i in range(plt.cm.Set2.N)]
        if len(colors) < len(data):
           raise ValueError('Automatic color assigment works for up to ' + str(len(colors)) + '.')
    if bottoms is None:
        bottoms = np.zeros(len(data))
    ticks_dist = (barh + bar_dist) * len(data) - bar_dist + cols_dist
    y = np.array([i * ticks_dist for i in range(len(data[0]))])
    #plt.xlim(0, np.max(data) + np.max(bottoms) + 1)
    ax = plt.subplot(111)
    barplots = []
    for i in range(len(data)):
        barplots.append(
            plt.barh(y + i * (barh + bar_dist), data[i], height=barh, left=bottoms[i], 
                     color=colors[i], edgecolor=colors[i], linewidth=1)
        )
        for di, val in enumerate(data[i]):
            ax.text(bottoms[i] + val + .25, y[di] + i * (barh + bar_dist), '{:.3f}'.format(val))
    ax.set_yticks(y + ((ticks_dist - cols_dist)/2 - barh/2))
    ax.set_yticklabels(labels)
    ax.legend(barplots, series_names)
    ax.invert_yaxis()
    ax.set_xlim(0, np.max(data) + np.max(bottoms) + 1.5)
    ax.set_xticks(np.arange(0, np.max(data) + np.max(bottoms) + 1, 1))
    ax.grid(axis='x', linewidth=.5)
    return ax
        
```

```python
def plot_models_feauture_importance(models, X, y, models_names=None, columns_labels=None, samples=1):
    if models_names is None:
        models_names = ['Model ' + str(i + 1) for i in range(len(models))]
    if columns_labels is None:
        columns_labels = list(X.columns)
    importance = []
    losses = []
    for model in models:
        loss, imp = features_importance(model, X, y, samples)
        losses.append(loss)
        importance.append(imp)
    plt.figure(figsize=(15,10))
    plt.suptitle('Feauters importance in different models', fontsize=18)
    ax = multiseries_horizontal_barchart(columns_labels, importance, models_names, losses)
    plt.plot()
```

### ALE plot dla wieku

Przyjrzymy się bliżej różnicom między modelom porównójąc ich ALE-ploty dla zmiennej wieku.
